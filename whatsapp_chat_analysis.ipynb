```python
import getpass, os
from langchain.prompts import PromptTemplate
from langchain.llms import OpenAI
from langchain.chains import LLMChain

# Get API key
os.environ['OPENAI_API_KEY'] = getpass.getpass("Enter your OpenAI API key:")

# Define sentiment analysis prompt
template = """
You are a helpful assistant that analyzes WhatsApp messages for tone and sentiment.
Given a message, classify it as Positive, Negative, or Neutral with a brief explanation.

Message: "{message}"
Sentiment:
"""
prompt = PromptTemplate(input_variables=["message"], template=template)
llm = OpenAI(temperature=0)
chain = LLMChain(llm=llm, prompt=prompt)

# Load messages from file
filename = '/mnt/data/file-7QJ31pMRA8tasvc2WPBEbs'  # Update this if needed
messages = load_whatsapp_chat('/content/whatsapp_chat_analysis.zip')

# Analyze sentiment
for i, msg in enumerate(messages[:5]):  # Analyze first 5 messages as a demo
    print(f"\nMessage {i+1}: {msg}")
    result = chain.run(message=msg)
    print(f"Sentiment Analysis: {result}")
```

```python
results = []

for item in messages[:10]:  # Limit to first 10 for cost
    try:
        response = chain.run({"message": item["message"]})
        results.append({
            "sender": item["sender"],
            "message": item["message"],
            "analysis": response.strip()
        })
    except Exception as e:
        results.append({
            "sender": item["sender"],
            "message": item["message"],
            "analysis": f"Error: {str(e)}"
        })
```
